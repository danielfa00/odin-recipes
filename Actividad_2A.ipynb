# **2. ¿Cuáles son los tipos de objetos que maneja la libería? Descríbelos utilizando los tipos básicos de Python (cadenas, números, listas, diccionarios y tuplas) cuando sea posible**






BeautifulSoup transforma un documento HTML en 4 tipos de objetos: Tag, NavigableString, BeautifulSoup y Comment.

**Tag**

Corresponde con un elemento HTML del documento con el que estamos trabajando.

Por ejemplo, si quiero navegar por la elemento body de mi HTML, puedo hacerlo así:

````python
# voy cambiando entre distintas etiquetas

a = soup.body # Todo el elemento
b = soup.body.name # Solo el nombre de la etiqueta
c = soup.title.string # Solo el contenido
d = soup.p.attrs # los atributos del elemento

print(a, b, c, d, sep="\n")
````

````python
# también podremos hacerlo así

title = soup.title

print(title.name)

# etc.
````

Veámoslo ahora en relación con los tipos básicos de Python.

````python
p = soup.p

p_nombre = p.name
p_contenido = p.string
p_atributos = p.attrs

print(p)
print(type(p))
print("\n")
print(p_nombre)
print(type(p_nombre))
print("\n")
print(p_contenido)
print(type(p_contenido))
print("\n")
print(p_atributos)
print(type(p_atributos))
````

De este modo, podemos decir, con carácter general, lo siguiente:

```soup.tag``` es un elemento especial propio de este módulo. Contiene todo el elemento HTML y tiene una codificación especial.

```soup.tag.name``` es una cadena.

```soup.tag.string``` curiosamente no es una cadena, sino un elemento especial de este módulo del que después hablaremos.

```soup.tag.attrs``` es un diccionario. El value es una lista, que en este caso está formada por un solo elemento.

Veamos qué pasa si navegamos un poco por un elemento Tag usando el atributo children:

````python
print(soup.html.children) # es un list iterator, así que usamos un loop

for item in soup.html.children:
  print("ITEM", item)

# vemos que por un lado tenemos <head> y por <body>

for item in soup.html.children:
  print(type(item))

# Cada uno de los elementos hijos es un objeto Tag (también hay dos NavigableString y no sé por qué)
````

Con ```.children``` y ```.contents``` solo obtenemos los hijos directos:

Mientras que con ```.children``` obtenemos un iterador, con ```.contents``` obtenemos una lista:

````python
z = soup.body.contents
print(type(z))

for item in z:
  print("ITEM", item)
````

````python
for item in soup.body.children:
  print("ITEM", item)
````

Una vez obtenemos los elementos Tag correspondientes (uno por cada elemento hijo directo), podemos hacer con ellos lo que ya hemos explicado:

````python
for item in soup.html.children:
  if item.name != None:  # Omitimos los NavigableString dentro del HTML
    print(f"Contenido: {item}\n")
    print(f"Tipo de elemento: {type(item)}\n")
    print(f"Nombre de la etiqueta: {item.name}\n")
    print(f"Tipo de elemento de item.name: {type(item.name)}") # Ya hemos visto antes que item.name es un string
    print("*********")
````

Para hacer una iteración por todos los elementos hijo de un elemento, usamos ```.descendants```:

````python
for item in soup.html.descendants:
  print(item)
````

**NavigableString**

BeautifulSoup usa la clase NavigableString para contener el texto que contiene un elemento HTML (o XML, claro). Lo hemos visto antes con ```p.string```.


Es parecido a un string normal de Python, pero tiene ciertas funcionalidades añadidas. Por ejemplo, contiene información sobre su ubicación dentro de la estructura del documento.  

Podemos transformarlo en una cadena normal y corriente usando ```unicode()```.

**BeautifulSoup**

Esta clase se usa para contener todo el documento HTML como un todo. Recordemos que antes hemos guardado todo el documento HTML en la variable ```soup``` por medio de la función BeautifulSoup(), que se usa para crear este tipo de objetos. Veámoslo:

````python
print(type(soup))
````

Si imprimimos ```soup``` nos devolverá todo el documento HTML, pero, como hemos visto, no se trata de un string.

Si queremos convertir un objeto BeautifulSoup como una cadena, podemos hacerlo así:

````python
soup_string = str(soup)
print(soup_string)
print(type(soup_string))
````

Observemos lo que pasa cuando con un objeto de tipo BeautifulSoup usamos el atributo .children:

````python
for item in soup.children:
  print("item", item)
  print(type(item))
````

El resultado que obtenemos es que todo el HTML está contenido en un elemento Tag que corresponde con la raíz ```<html>```.



**Comment**

Es un tipo especial de NavigableString, que se corresponde con los comentarios del HTML ```<!-- ... -->```.

# **7. ¿Es posible filtrar una lista de elementos? Si es posible, escribe el código necesario para extraer los elementos *p* y *table* de un documento HTML.**

Sí, es posible, y para ello existen distintos procedimientos.

````python
html_simple = """
<html>
<head>
<title>HTML simple</title>
</head>
<body>
<p>Párrafo de prueba</p>
<p id="p1">Otro</p>
<h3>Heading de prueba</h3>
<table style="border: solid black 1px; border-collapse: collapse">
<tr>
<th>País</th>
<th>Capital</th>
</tr>
<tr>
<td>España</td>
<td>Madrid</td>
</tr>
</table>
<h4>Otro heading</h4>
<p>Otro ejemplo</p>
<p>Este ya es el último párrafo</p>
</body>
</html>
"""
````

````python
soup4 = BeautifulSoup(html_simple, "html.parser")
````

Filtramos el HTML para obtener todos los elementos ```<p>``` y obtenemos algo parecido a una lista:

````python
resultados_p = soup4.find_all("p")

print(resultados_p)
print(type(resultados_p))
````

Con este objeto iterable podemos hacer cosas:

````python
for item in resultados_p:
  print(item.string)
````

Extraigamos ahora la tabla:

````python
resultados_tabla = soup4.find_all("table")
print(resultados_tabla)
````
